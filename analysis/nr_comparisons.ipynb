```python
import random
import numpy as np
import matplotlib.pyplot as plt 
import os
```

```python
def merge_sort(A):    
    if len(A) > 1:
         #increment
        comparisons.append(1)
        
        middle_item = len(A)//2
        right_split = A[middle_item:]
        left_split = A[:middle_item]
        merge_sort(left_split)
        merge_sort(right_split)
        A = merge(A, left_split, right_split) 
        
    return A

def merge(A, left_split, right_split, start=0):   
    i, j, k = 0, 0, start
        
    while i < len(left_split) and j < len(right_split):
        
        #increment
        comparisons.append(2)
        
        if left_split[i] <= right_split[j]:
            A[k] = left_split[i] 
            i += 1
            
        else:
            A[k] = right_split[j] 
            j += 1
        k += 1
        
        #increment
        comparisons.append(1)

        
        
    while i < len(left_split): 
        A[k] = left_split[i] 
        k += 1
        i += 1
        
        #increment
        comparisons.append(1)
  
    while j < len(right_split): 
        A[k] = right_split[j] 
        k += 1
        j += 1
        
        #increment
        comparisons.append(1)
```

```python
np.random.seed(12235)
comparison_plots = []
list_length =[2**i for i in range(4, 15)]
for i in list_length: 
    #initilize
    comparisons = []
    gen_data = np.random.uniform(size=i)
    test_list = sorted(gen_data, reverse=True)
    sorted_data = merge_sort(test_list)
    comparison_plots.append(sum(comparisons))
    #initilize
    comparisons = [] 
    test_list = sorted(gen_data)
    sorted_data = merge_sort(test_list)
    comparison_plots.append(sum(comparisons))
    #initilize
    comparisons = [] 
    random.shuffle(gen_data)
    sorted_data = merge_sort(gen_data)
    comparison_plots.append(sum(comparisons))
comparison_plots = np.array(comparison_plots).reshape(11, 3)
```

```python
print('Merge sort on reverse data of size 1024:',comparison_plots[6][0],
     '\nMerge sort on sorted data of size 1024:',comparison_plots[6][1],  
     '\nMerge sort on random data of size 1024:',comparison_plots[6][2])
print('Merge sort on reverse data of size 2048:',comparison_plots[7][0],  
     '\nMerge sort on sorted data of size 2048:',comparison_plots[7][1], 
     '\nMerge sort on random data of size 2048:',comparison_plots[7][2])
```

```python
x = np.arange(comparison_plots.shape[0])
dx = (np.arange(comparison_plots.shape[1])-comparison_plots.shape[1]/2.)/(comparison_plots.shape[1]+2.)
d = 1./(comparison_plots.shape[1]+2.)
#fig, ax=plt.subplots()
plt.figure(figsize=(25,15))

for i in range(comparison_plots.shape[1]):
    plt.bar(x+dx[i],comparison_plots[:,i], width=d, tick_label=list_length)
plt.legend(['reversed', 'sorted','random'], fontsize=42)
plt.xlabel('number of elements', fontsize=50)
plt.ylabel('number of comparisons', fontsize=50)
plt.title('a', fontsize=50)
plt.xticks(fontsize=50, rotation='60', ha='right')
plt.yticks(fontsize=50)
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
plt.rc('font', size=50)
plt.tick_params(width=5,length=10)
plt.grid()
plt.tight_layout()
plt.savefig('../figures/merge_number_comparisons.pdf')
plt.show()
```

```python
def quick_sort(A):
    left_part = []
    equal_part = []
    right_part = []

    
    if len(A) > 1:
        pivot = A[len(A)//2]
        for element in A:
            if element < pivot:
                comparisons.append(1) 
                left_part.append(element)
            elif element == pivot:
                comparisons.append(1)  
                equal_part.append(element)
            elif element > pivot:
                comparisons.append(1)  
                right_part.append(element)
            #initilize
        A = quick_sort(left_part)+equal_part+quick_sort(right_part)
    return A
```

```python
np.random.seed(12235)
comparison_plots = []
list_length =[2**i for i in range(4, 15)]
for i in list_length: 
    #initilize
    comparisons = []
    gen_data = np.random.uniform(size=i)
    test_list = sorted(gen_data, reverse=True)
    sorted_data = quick_sort(test_list)
    comparison_plots.append(sum(comparisons))
    #initilize
    comparisons = [] 
    test_list = sorted(gen_data)
    sorted_data = quick_sort(test_list)
    comparison_plots.append(sum(comparisons))
    #initilize
    comparisons = [] 
    random.shuffle(gen_data)
    sorted_data = quick_sort(gen_data)
    comparison_plots.append(sum(comparisons))
comparison_plots = np.array(comparison_plots).reshape(11, 3)
```
