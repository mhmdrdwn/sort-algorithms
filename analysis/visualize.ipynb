### Quadratic algorithms Comparision

```python
## quadratic sort statistical analysis 
yy=['bubble_sorted', 'insertion_sorted','bubble_reverse', 'insertion_reverse','bubble_random', 'insertion_random']
stat_quad=df[yy]
stat_quad.describe()
```

```python
from scipy.optimize import curve_fit
def quad_func(x, a):
    return a * (x**2)
def quad_curve_fit(x, y):
    popt, pcov = curve_fit(quad_func, x, y)
    return popt

y= ['bubble_random', 'insertion_random']
x = 'list_length'
plt.figure(figsize=(25,15))
toplot = np.arange(1,16400)

for i, col in enumerate(y):
    plt.plot(df[x], df[col], linestyle = ' ', marker='s', ms=15, color=colors[i])
    popt = quad_curve_fit(df[x], df[col])
    print(popt)
    plt.plot(toplot, quad_func(toplot, *popt), linestyle = '--', linewidth=4, alpha=0.7, color=colors[i])
    plt.xlabel('number of elements', fontsize=50)
    plt.ylabel('run time [seconds]', fontsize=50)
    plt.title('a', fontsize=50)
    plt.legend(['bubble', 'fit curve: $7.3e^{-7}x^{2}$', 'insertion', ' fit curve: $1.9e^{-7}x^{2}$'], fontsize=42)
    plt.xticks(fontsize=50)
    plt.yticks(fontsize=50)
    plt.tick_params(width=5,length=10)
plt.grid()
plt.tight_layout()
plt.savefig('../figures/bubble_vs_insertion_random.pdf')
plt.show()
```

```python
y= ['bubble_sorted', 'insertion_sorted']
x = 'list_length'

def linear_func(x, a):
    return a * x
def linear_curve_fit(x, y):
    popt, pcov = curve_fit(linear_func, x, y)
    return popt

plt.figure(figsize=(25,15))
toplot = np.arange(1,16400)
for i, col in enumerate(y):
    plt.plot(df[x], df[col], linestyle = ' ', marker='s', ms=15, color=colors[i])
    if col == 'insertion_sorted':
        popt = linear_curve_fit(df[x], df[col])
        plt.plot(toplot, linear_func(toplot, *popt), linestyle = '--', linewidth=4, alpha=0.7, color=colors[i])
    else:
        popt = quad_curve_fit(df[x], df[col])
        plt.plot(toplot, quad_func(toplot, *popt), linestyle = '--', linewidth=4, alpha=0.7, color=colors[i])
    print(popt)
    plt.xlabel('number of elements', fontsize=50)
    plt.ylabel('run time [seconds]', fontsize=50)
    plt.title('b', fontsize=50)
    plt.legend(['bubble', 'fit curve: $2e^{-7}x^{2}$', 'insertion', 'fit curve: $3.6e^{-7}x$'], fontsize=42)
    plt.xticks(fontsize=50)
    plt.yticks(fontsize=50)
    plt.tick_params(width=5,length=10)
plt.grid()
plt.tight_layout()
plt.savefig('../figures/bubble_vs_insertion_sorted.pdf')
plt.show()
```

```python
y= ['bubble_reverse', 'insertion_reverse']
x = 'list_length'

plt.figure(figsize=(25,15))
toplot = np.arange(1,16400)

for i, col in enumerate(y):
    plt.plot(df[x], df[col], linestyle = ' ', marker='s', ms=15, color=colors[i])
    popt = quad_curve_fit(df[x], df[col])
    print(popt)
    plt.plot(toplot, quad_func(toplot, *popt), linestyle = '--', linewidth=4, alpha=0.7, color=colors[i])
    plt.xlabel('number of elements', fontsize=50)
    plt.ylabel('run time [seconds]', fontsize=50)
    plt.title('c', fontsize=50)
    plt.legend(['bubble', 'fit curve: $3.6e^{-7}x^{2}$', 'insertion', 'fit curve: $1.6e^{-7}x^{2}$'], fontsize=42)
    plt.xticks(fontsize=50)
    plt.yticks(fontsize=50)
    plt.tick_params(width=5,length=10)
plt.grid()
plt.tight_layout()
plt.savefig('../figures/bubble_vs_insertion_reverse.pdf')
plt.show()
```

### Sub-quadratic

```python
def log_func(x, a, b):
    return a * x * (np.log2(x))**b
def log_curve_fit(x, y):
    popt, pcov = curve_fit(log_func, x, y)
    return popt


y= ['quick_random','merge_random','combined_random']
x = 'list_length'

plt.figure(figsize=(25,15))
toplot = np.arange(1,16400)

for i, col in enumerate(y):
    plt.plot(df[x], df[col], linestyle = ' ', marker='s', ms=15, color=colors[i])
    popt = log_curve_fit(df[x], df[col])
    print(popt)
    plt.plot(toplot, log_func(toplot, *popt), linestyle = '--', linewidth=4, alpha=0.7, color=colors[i])
    plt.xlabel('number of elements', fontsize=50)
    plt.ylabel('run time [seconds]', fontsize=50)
    plt.title('a', fontsize=50)
    plt.legend(['quick', 'fit curve: $5.3e^{-7}n.\lg(n)^{0.9}$', 'merge', 'fit curve: $1.4e^{-6}n.\lg(n)^{0.8}$', 'combined', 'fit curve: $1.3e^{-6}n.\lg(n)^{0.9}$'], fontsize=42)
    plt.xticks(fontsize=50)
    plt.yticks(fontsize=50)
    plt.tick_params(width=5,length=10)
    plt.grid()
    
plt.tight_layout()
plt.savefig('../figures/sub_sorts_random.pdf')
plt.show()
```

# Boxplots

```python
all_data = pd.read_csv('../data/all_data.csv', delimiter = ',')
all_data
```

```python
random_df = all_data[all_data.case=='random_case']
sorted_df = all_data[all_data.case=='sorted_case']
reverse_df = all_data[all_data.case=='reverse_case']
```

```python
df_size = pd.DataFrame( random_df[random_df['list_length']==16384])
numpy = df_size[(df_size['sort_type']=='numpy_sort')] ['run_time'] 
python = df_size[(df_size['sort_type']=='python_sort')] ['run_time']  
quick = df_size[(df_size['sort_type']=='quick_sort')] ['run_time']  
merge = df_size[(df_size['sort_type']=='merge_sort')] ['run_time']  
insert = df_size[(df_size['sort_type']=='insertion_sort')] ['run_time'] 
bubble = df_size[(df_size['sort_type']=='bubble_sort')] ['run_time']
combined = df_size[(df_size['sort_type']=='combined_sort')] ['run_time']

data=[numpy, python,quick,merge, insert, bubble, combined]
labels = ['numpy', 'python', 'quick', 'merge', 'insertion', 'bubble', 'combined']
for i, j in zip(data, labels):
    plt.figure(figsize=(9,15))
    c = 'lightblue'
    plt.boxplot(i)
    if j is 'bubble':
        plt.ylabel('run time [seconds]', fontsize=70)
    plt.yticks(fontsize=70)
    xlabels = [j]
    locs, labels=plt.xticks()
    plt.xticks(locs, xlabels, fontsize=70)
    plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
    plt.rc('font', size=70)
    plt.tick_params(width=5,length=10)
    plt.tight_layout()
    plt.savefig('../figures/all_random_'+j+'_box.pdf')
plt.show()
```
